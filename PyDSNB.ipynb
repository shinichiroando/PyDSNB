Import updated PyDSNB code and set plot defaults.

```python
from PyDSNB import *

plt.rc('font', **{'family':'sans-serif','size':15})
plt.rc('text', usetex=False)
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rcParams['xtick.major.size']  = 5
plt.rcParams['xtick.major.width'] = 1
plt.rcParams['xtick.minor.size']  = 2
plt.rcParams['xtick.minor.width'] = 1
plt.rcParams['ytick.major.size']  = 5
plt.rcParams['ytick.major.width'] = 1
plt.rcParams['ytick.minor.size']  = 2
plt.rcParams['ytick.minor.width'] = 1
plt.rcParams['xtick.direction']   = 'in'
plt.rcParams['ytick.direction']   = 'in'
plt.rcParams['xtick.major.pad']   = 8
plt.rcParams['xtick.top']         = True
plt.rcParams['ytick.right']       = True
```

Read the DSNB class with an initialization file (in this example 'params.ini').

```python
sn = DSNB('initialization/sample/params.ini')
```

Star-formation rate density as a function of redshift.

```python
z = np.linspace(0,5,1000)
plt.semilogy(z,sn.R_SF(z)/(sn.Msun/sn.yr/sn.Mpc**3))
plt.xlabel(r'Redshift')
plt.ylabel(r'$R_{\rm SN}(z) \ [\mathrm{M_{\odot}\, yr^{-1}\, Mpc^{-3}}]$')
plt.show()
```

Supernova neutrino spectrum.

```python
E = np.linspace(0.1,50)*sn.MeV
plt.semilogy(E/sn.MeV,sn.dNdE_func(E)/(sn.MeV**-1),label=sn.SNmodel)
plt.xlabel(r'$E$ [MeV]')
plt.ylabel(r'$dN_{\bar\nu_e}/dE$ [$\mathrm{MeV^{-1}}$]')
plt.legend()
plt.show()
```

DSNB spectrum.

```python
E = np.linspace(0.,100.,1001)*sn.MeV
dFdE = sn.DSNB_dFdE(E)

plt.xlim(0.1,50)
plt.ylim(1e-5,10)
plt.semilogy(E/sn.MeV,dFdE/(sn.cm**-2*sn.s**-1*sn.MeV**-1),label=sn.SNmodel)
plt.xlabel(r'$E$ [MeV]')
plt.ylabel(r'$dF_{\bar\nu_e}/dE$ [$\mathrm{cm^{-1}\,s^{-1}\,MeV^{-1}}$]')
plt.legend()
plt.show()
```

Compare 13M, 26M, and IMF weighed DSNB spectra.

```python
sn2 = DSNB('initialization/sample/params2.ini')
sn3 = DSNB('initialization/sample/params3.ini')
```

```python
E = np.linspace(0.,100.,1001)*sn.MeV
dFdE = sn.DSNB_dFdE(E)
dFdE2 = sn2.DSNB_dFdE(E)
dFdE3 = sn3.DSNB_dFdE(E)

plt.xlim(0.1,50)
plt.ylim(1e-5,10)
plt.semilogy(E/sn.MeV,dFdE/(sn.cm**-2*sn.s**-1*sn.MeV**-1),label='13M')
plt.semilogy(E/sn2.MeV,dFdE2/(sn2.cm**-2*sn2.s**-1*sn2.MeV**-1),label='26M')
plt.semilogy(E/sn3.MeV,dFdE3/(sn3.cm**-2*sn3.s**-1*sn3.MeV**-1),label='IMF')
plt.xlabel(r'$E$ [MeV]')
plt.ylabel(r'$dF_{\bar\nu_e}/dE$ [$\mathrm{cm^{-1}\,s^{-1}\,MeV^{-1}}$]')
plt.legend()
plt.show()
```

Compare IMF weighed spectra of other CCSN models.

```python
sn4 = DSNB('initialization/sample/params4.ini')
sn5 = DSNB('initialization/sample/params5.ini')
sn6 = DSNB('initialization/sample/params6.ini')
```

```python
E = np.linspace(0.,100.,1001)*sn.MeV
dFdE3 = sn3.DSNB_dFdE(E)
dFdE4 = sn4.DSNB_dFdE(E)
dFdE5 = sn5.DSNB_dFdE(E)
dFdE6 = sn6.DSNB_dFdE(E)

plt.xlim(0.1,50)
plt.ylim(1e-5,10)
plt.semilogy(E/sn3.MeV,dFdE3/(sn3.cm**-2*sn3.s**-1*sn3.MeV**-1),label=sn3.SNmodel)
plt.semilogy(E/sn4.MeV,dFdE4/(sn4.cm**-2*sn4.s**-1*sn4.MeV**-1),label=sn4.SNmodel)
plt.semilogy(E/sn5.MeV,dFdE5/(sn5.cm**-2*sn5.s**-1*sn5.MeV**-1),label=sn5.SNmodel)
plt.semilogy(E/sn6.MeV,dFdE6/(sn6.cm**-2*sn6.s**-1*sn6.MeV**-1),label=sn6.SNmodel)
plt.xlabel(r'$E$ [MeV]')
plt.ylabel(r'$dF_{\bar\nu_e}/dE$ [$\mathrm{cm^{-1}\,s^{-1}\,MeV^{-1}}$]')
plt.legend()
plt.show()
```

Estimate the late phase neutrino emission with the correlation method (estimating the shock revival time and final PNS baryon mass).

```python
sn7 = DSNB('initialization/sample/params7.ini')
```

```python
E = np.linspace(0.,100.,1001)*sn.MeV
dFdE3 = sn3.DSNB_dFdE(E)
dFdE7 = sn7.DSNB_dFdE(E)

plt.xlim(0.1,50)
plt.ylim(1e-5,10)
plt.semilogy(E/sn3.MeV,dFdE3/(sn3.cm**-2*sn3.s**-1*sn3.MeV**-1),label=sn3.SNmodel)
plt.semilogy(E/sn7.MeV,dFdE7/(sn7.cm**-2*sn7.s**-1*sn7.MeV**-1),label='F21 + late phase')
plt.xlabel(r'$E$ [MeV]')
plt.ylabel(r'$dF_{\bar\nu_e}/dE$ [$\mathrm{cm^{-1}\,s^{-1}\,MeV^{-1}}$]')
plt.legend()
plt.show()
```

Compare late phase estimations with four different methods.

```python
sn8 = DSNB('initialization/sample/params8.ini')
sn9 = DSNB('initialization/sample/params9.ini')
sn10 = DSNB('initialization/sample/params10.ini')
```

```python
E = np.linspace(0.,100.,1001)*sn.MeV
dFdE7 = sn7.DSNB_dFdE(E)
dFdE8 = sn8.DSNB_dFdE(E)
dFdE9 = sn9.DSNB_dFdE(E)
dFdE10 = sn10.DSNB_dFdE(E)

plt.xlim(0.1,50)
plt.ylim(1e-5,10)
plt.semilogy(E/sn7.MeV,dFdE7/(sn7.cm**-2*sn7.s**-1*sn7.MeV**-1),label='F21 + Corr')
plt.semilogy(E/sn8.MeV,dFdE8/(sn8.cm**-2*sn8.s**-1*sn8.MeV**-1),label='F21 + RenormShen')
plt.semilogy(E/sn9.MeV,dFdE9/(sn9.cm**-2*sn9.s**-1*sn9.MeV**-1),label='F21 + RenormLS')
plt.semilogy(E/sn10.MeV,dFdE10/(sn10.cm**-2*sn10.s**-1*sn10.MeV**-1),label='F21 + Analyt')
plt.xlabel(r'$E$ [MeV]')
plt.ylabel(r'$dF_{\bar\nu_e}/dE$ [$\mathrm{cm^{-1}\,s^{-1}\,MeV^{-1}}$]')
plt.legend()
plt.show()
```
